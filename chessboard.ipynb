```python
# TaskPY200_T5: Chessboard with Numbers

N = int(input("Enter chessboard size: "))

for row in range(N):
    for col in range(N):
        if row == col:
            print("X", end=" ")
        elif (row + col) % 2 == 0:
            print("1", end=" ")
        else:
            print("0", end=" ")
    print()
```
